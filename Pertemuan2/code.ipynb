Import perpustakaan

```python
import pandas as pd
import numpy as np
import sqlalchemy as db
import matplotlib.pyplot as plt
import seaborn as sns
```

Panggil database data_hujan.sqlite

```python
database = db.create_engine("sqlite:///data_hujan_fix.sqlite")
conn = database.connect()
```

```python
# Tanpa CTE
data_gabungan_1 = """
    SELECT
        a1.Wilayah,
        a1.CH,
        a2.Temp
    FROM data_hujan a1
    LEFT JOIN data_temperatur a2 ON a1.CH = a2.CH
"""

data_bersih_no_CTE = pd.read_sql_query(data_gabungan_1, conn)
data_bersih_no_CTE
```

```python
# Menggunakan CTE
data_gabungan_2 = """
    WITH data_x AS (
        SELECT
            a1.Wilayah,
            a1.CH,
            a2.Temp
        FROM data_hujan a1
        LEFT JOIN data_temperatur a2 ON a1.CH = a2.CH
    )

    SELECT * FROM data_x
"""

data_CTE = pd.read_sql_query(data_gabungan_2, conn)
data_CTE
```

Kita mencoba panggil data darai SQLdatabasenya dan gabungan data menjadi 1

```python
data_CTE.to_csv("data_gabungan_final.csv", index=None) # Gabungkan data
```

```python
data_curah_hujan = data_CTE.groupby('Wilayah')
data_curah_hujan_avg = data_curah_hujan.agg(
    rata_rata = ("CH","mean")
)
data_curah_hujan_avg_fix = data_curah_hujan_avg.reset_index()
data_curah_hujan_avg_fix
```

```python
data_curah_hujan_max = data_CTE.groupby("Wilayah")
data_curah_hujan_sum = data_curah_hujan_max.agg(
    jumlah = ("CH", "sum")
)

data_curah_hujan_sum = data_curah_hujan_sum.reset_index()
data_curah_hujan_sum.loc[data_curah_hujan_sum["Wilayah"].idxmax()]
```
